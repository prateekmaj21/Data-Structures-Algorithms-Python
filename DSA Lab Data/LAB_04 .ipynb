**No. 06 - Practice Problems: Linked Lists  **

Implementation of the “Single Linked List” Algorithm to create a program in C/Python. Perform the following operations:

1.Insert at beginning  
2.Insert at end  
3.Insert at any position
4.Delete at beginning  
5.Delete at end  
6.Delete at any position





```python
class Node:
    def __init__(self, data):
        self.data = data  # Initialize node with data and next pointer
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None  # Initialize an empty linked list with no head

    def insertBegin(self, new_node):
        if self.head is None:  # If linked list is empty, new node becomes head
            self.head = new_node
        else:
            current = self.head
            new_node.next = current  # Insert new node at the beginning
            self.head = new_node

    def insertEnd(self, new_node):
        if self.head is None:  # If linked list is empty, new node becomes head
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node  # Insert new node at the end


    def insertAtAnyPos(self, pos, new_node):
       if self.head is None and pos == 1:
        new_node.next = self.head
        self.head = new_node
       else:
        flag = False  # Flag to track if the position is found
        current = self.head  # Start traversal from the head of the list
        prev = None  # Initialize previous node to None
        current_pos = 1  # Variable to track the current position in the list

        # Traverse the list until current reaches the last node
        while current is not None:
            # Check if the target position is the first node
            if pos == 1:
                self.insertBegin(new_node)  # Call insertBegin method to insert at the beginning
                flag = True  # Set flag to True to indicate insertion
                break  # Exit the loop
            # Check if the current position matches the target position
            elif current_pos == pos:
                prev.next = new_node  # Update pointers to insert the new node at target position
                new_node.next = current
                flag = True  # Set flag to True to indicate insertion
                break  # Exit the loop
            else:
                # Move to the next node in the list
                current_pos += 1
                prev = current
                current = current.next

        # Check if the target position is found
        if flag == False and current_pos == pos:
            prev.next = new_node  # Update pointers to insert the new node at the end
            flag = True  # Set flag to True to indicate insertion
        elif flag == True:
            pass  # If the position is found and new node inserted, do nothing
        else:
            print("Invalid position")  # Print message for invalid position



    def deleteBegin(self):
        if self.head is None:
            print("Can't delete, Linked list is empty")
        else:
            self.head = self.head.next  # Delete node from the beginning

    def deleteEnd(self):
        if self.head is None:
            print("Can't delete, Linked list is empty")
        else:
            current = self.head
            prev = current
            while current.next is not None:
                prev = current
                current = current.next
            if self.head.next is None:#only one node in list
                self.head = None
            else:
                prev.next = None  # Delete node from the end

    def deleteAtAnyPos(self, pos):
       if self.head is None:
        print("Can't delete, Linked list is empty")
       else:
        flag = False  # Flag to track if the position is found
        current = self.head  # Start traversal from the head of the list
        prev = current  # Initialize previous node to current for the first iteration
        current_pos = 1  # Variable to track the current position in the list

        # Traverse the list until current reaches the last node
        while current is not None:
            # Check if the target position is the first node
            if pos == 1:
                self.deleteBegin()  # Call deleteBegin method to delete the first node
                flag = True  # Set flag to True to indicate deletion
                break  # Exit the loop
            # Check if the current position matches the target position
            elif current_pos == pos:
                prev.next = current.next  # Update pointers to remove the node at target position
                flag = True  # Set flag to True to indicate deletion
                break  # Exit the loop
            else:
                # Move to the next node in the list
                current_pos += 1
                prev = current
                current = current.next

        # Check if the target position is found
        if flag == False:
            print("Invalid position")  # Print message for invalid position


    def printLL(self):
      ll = ''  # Initialize an empty string to store the linked list representation
      current = self.head  # Start traversal from the head of the list

    # Traverse the list until current reaches the last node (current becomes None)
      while current:
          data = current.data  # Get the data stored in the current node
          ll = ll + f'{data} -->'  # Append the data to the linked list representation
          current = current.next  # Move to the next node in the list

      ll += 'NULL'  # Add 'NULL' to indicate the end of the linked list
      print("Linked List: ", ll)  # Print the linked list representation


if __name__ == '__main__':
    ll = LinkedList()
    while True:
        user_input = int(input("""Enter 1 to insert at beginning
Enter 2 to insert at end
Enter 3 to insert at any position
Enter 4 to delete at beginning
Enter 5 to delete at end
Enter 6 to delete at any position
Enter 7 to display the Linked list
Enter any number to exit\n"""))
        if user_input == 1:
            data = int(input("Enter the node data: "))
            new_node = Node(data)
            ll.insertBegin(new_node)
        elif user_input == 2:
            data = int(input("Enter the node data: "))
            new_node = Node(data)
            ll.insertEnd(new_node)
        elif user_input == 3:
            data = int(input("Enter the node data: "))
            new_node = Node(data)
            pos = int(input("Enter the position: "))
            ll.insertAtAnyPos(pos, new_node)
        elif user_input == 4:
            ll.deleteBegin()
        elif user_input == 5:
            ll.deleteEnd()
        elif user_input == 6:
            pos = int(input("Enter the position: "))
            ll.deleteAtAnyPos(pos)
        elif user_input == 7:
            ll.printLL()
        else:
            break
```

**No. 07 - Practice Problems: Doubly and Circular Linked Lists  **

Insert into (at beginning, at any position and at end) and delete (beginning, at any position and at end) from a Doubly Linked List.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList():
    def __init__(self):
        self.head = None

    # Function to insert a new node at the beginning
    def insertBegin(self, new_node):
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insertEnd(self, new_node):
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def insertAtAnyPos(self, new_node, pos):
        if self.head is None and pos == 1:
            self.head = new_node
        else:
            flag = False
            current_pos = 1
            current = self.head
            previous = None
            while current is not None:
                if pos == 1:
                    self.insertBegin(new_node)
                    flag = True
                    break
                elif current_pos == pos:
                    previous.next = new_node
                    new_node.prev = previous
                    new_node.next = current
                    flag = True
                    break
                else:
                    current_pos+=1
                    previous = current
                    current = current.next
            if flag == False and current_pos == pos:
                previous.next = new_node
                new_node.prev = previous
            elif flag == True:
                pass
            else:
                print("Invalid position")

    def deleteBegin(self):
        if self.head is None:
            print("Can't delete, Linked list is empty")
        else:
            current = self.head
            self.head = self.head.next #pointing to new head
            if current.next is not None:
                current.next.prev = None #Breaking the link to Head by next element(<--prev)
                current.next = None #Breaking the link of Head to next Element(--->next)
            else:
                self.head = None

    def deleteEnd(self):
      if self.head is None: # Check if the linked list is empty
        print("Can't delete, Linked list is empty")
      else:
        current = self.head  # Start traversal from the head of the list
        previous = None  # Initialize previous node to None

        # Traverse the list until current reaches the last node
        while current.next is not None:
            previous = current  # Update previous to the current node
            current = current.next  # Move current to the next node

        # After the loop, current is the last node

        # If the head is the only node in the list
        if self.head.next is None:
            self.head = None  # Set head to None, making the list empty

        # If there are multiple nodes in the list
        else:
            previous.next = None  # Disconnect the last node from the list
            current.prev = None  # Sever its connection to the previous node


    def deleteAtAnyPos(self, pos):
      if self.head is None: # Check if the linked list is empty
        print("Can't delete, Linked list is empty")
      else:
        flag = False  # Flag to track if the target position is found
        current_pos = 1  # Variable to track the current position in the list
        current = self.head  # Start traversal from the head of the list
        previous = current  # Initialize previous node to current

        # Traverse the list until current reaches the last node
        while current.next is not None:
            # Check if the target position is the first node
            if pos == 1:
                self.deleteBegin()  # Call deleteBegin method to delete the first node
                flag = True  # Set flag to True to indicate deletion
                break  # Exit the loop
            # Check if the current position matches the target position
            elif current_pos == pos:
                # Update pointers to remove the node at the target position
                current.prev = None
                previous.next = current.next
                current.next.prev = previous
                current.next = None
                flag = True  # Set flag to True to indicate deletion
                break  # Exit the loop
            else:
                # Move to the next node in the list
                current_pos += 1
                previous = current
                current = current.next

        # Check if the target position is found
        if flag == True:
            pass  # If the position is found and deleted, do nothing
        # Check if the target position is not found
        elif flag == False and current_pos == pos:
            # Handle deletion of the last node in the list
            if current_pos == 1:
                self.head = None
            current.prev = None
            previous.next = None
        else:
            # Print message for invalid position
            print("Invalid position")


    def printLL(self):
        temp = self.head
        ll=''
        while temp is not None:
            ll += f'{temp.data}' + '<->'
            temp= temp.next
        ll+='NULL'
        print("Linked List: ", ll)

# Code execution starts here
if __name__=='__main__':
    # Start with the empty list
    dll = DoublyLinkedList()

    # Insert nodes at the beginning
    new_node=Node(20)
    dll.insertBegin(new_node)
    new_node=Node(10)
    dll.insertBegin(new_node)
    dll.printLL()

    # Insert node at the end
    new_node=Node(40)
    dll.insertEnd(new_node)
    dll.printLL()

    # Insert node at position 4
    new_node = Node(100)
    dll.insertAtAnyPos(new_node,3)
    dll.printLL()

    # Delete node from the beginning
    dll.deleteBegin()
    dll.printLL()

    # Delete node from the end
    dll.deleteEnd()
    dll.printLL()

    # Delete node from position 2
    dll.deleteAtAnyPos(2)
    dll.printLL()

    # Delete node from position 1
    dll.deleteAtAnyPos(1)
    dll.printLL()

    # Delete node from position 1
    dll.deleteAtAnyPos(1)
    dll.printLL()
```

**No.  09 - Practice Problems: Tree and its Traversal  **

Implementation of the “Binary Tree” Algorithms to create a program in C/Python. Also, a program to implement “Pre Order, In Order, Post Order Traversal of Binary Tree (Recursive & Non-Recursive- Not mandatory)” Algorithms.

```python
# Define a class for a Node in a binary tree
class Node:
    def __init__(self, data):
        self.data = data  # Data stored in the node
        self.left = None  # Reference to the left child node
        self.right = None # Reference to the right child node

# Function to print the binary tree in level order (BFS)
def print_level_order(root):
    if not root:  # If the root is None, return
        return

    queue = [root]  # Initialize a queue with the root node
    while queue:
        current_level = [] # List to store the nodes at the current level
        next_level = []    # List to store the nodes at the next level

        for node in queue: # Iterate through the nodes in the current level
            current_level.append(node.data) # Add the data of the current node to the current level list
            if node.left:   # If the current node has a left child
                next_level.append(node.left) # Add the left child to the next level list
            if node.right:  # If the current node has a right child
                next_level.append(node.right) # Add the right child to the next level list

        print(" ".join(str(val) for val in current_level)) # Print the nodes' data at the current level
        queue = next_level  # Update the queue with nodes from the next level

# Function to print the binary tree in inorder traversal #order:left subtree, root, right subtree.
def printInorder(root):
    if root:
        # First recur on left child
        printInorder(root.left)
        # Then print the data of node
        print(root.data, end=" ")
        # Now recur on right child
        printInorder(root.right)

# Function to print the binary tree in preorder traversal #order:root, left subtree, right subtree.
def printPreorder(root):
    if root:
        # Then print the data of node
        print(root.data, end=" ")
        # First recur on left child
        printPreorder(root.left)
        # Now recur on right child
        printPreorder(root.right)

# Function to print the binary tree in postorder traversal  order: left subtree, right subtree, root.
def printPostorder(root):
    if root:
        # First recur on left child
        printPostorder(root.left)
        # Now recur on right child
        printPostorder(root.right)
        # Then print the data of node
        print(root.data, end=" ")

# Main block
if __name__ == '__main__':
    # Create the root node
    root = Node(1)
    # Create level 2 nodes
    l1 = Node(2)
    r1 = Node(3)
    # Connect the level 2 nodes to the root
    root.left = l1
    root.right = r1
    # Create level 3 nodes
    l21 = Node(4)
    r21 = Node(5)
    l22 = Node(6)
    r22 = Node(7)
    # Connect the level 3 nodes to their parent nodes
    l1.left = l21
    l1.right = r21
    r1.left = l22
    r1.right = r22
    # Print the binary tree in level order
    print_level_order(root)
    # Print the binary tree in inorder traversal
    print("Inorder: ")
    printInorder(root)
    # Print the binary tree in preorder traversal
    print("\nPreorder: ")
    printPreorder(root)
    # Print the binary tree in postorder traversal
    print("\nPostorder: ")
    printPostorder(root)
```
