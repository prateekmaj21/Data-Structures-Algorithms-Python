Implementation of the “Stack using an array”. Use these concept: Push(value), Pop(), FULL, EMPTY, and FindTOP position.

```python
# Function to push an element onto the stack
def Push(stack, value, stack_size):
    if len(stack) < stack_size:  # Check if the stack is not full
        stack.append(value)  # Add the value to the stack
    else:
        print("Stack Overflow")  # Print message if stack is full
    return stack

# Function to pop an element from the stack
def Pop(stack):
    if len(stack) == 0:  # Check if the stack is empty
        print("Stack Underflow")  # Print message if stack is empty
        return
    else:
        poped_ele = stack[-1]  # Get the top element(Last element)of the stack
        print("Popped element:", poped_ele)  # Print the popped element
        del stack[-1]  # Remove the top element from the stack
        return stack

# Function to check if the stack is full
def isFull(stack, stack_size):
    if len(stack) == stack_size:  # Check if the length of stack is equal to stack size
        print("The stack is full")  # Print message if stack is full
    else:
        print("The stack is not full")  # Print message if stack is not full

# Function to check if the stack is empty
def isEmpty(stack):
    if len(stack) == 0:  # Check if the length of stack is 0
        print("The stack is empty")  # Print message if stack is empty
    else:
        print("The stack is not empty")  # Print message if stack is not empty

# Function to find the top element of the stack
def findTop(stack):
    if len(stack) == 0:  # Check if the stack is empty
        print("The stack is empty")  # Print message if stack is empty
    else:
        print("The top element is:", stack[-1])  # Print the top element of the stack

# Main code
stack_size = int(input("Enter the size of the stack: "))  # Input the size of the stack
stack = []  # Initialize an empty stack

while True:
    user_input = int(input("""Enter 1 to push an element in the stack
Enter 2 to pop the element from the stack
Enter 3 to check whether the stack is full or not
Enter 4 to check whether the stack is empty or not
Enter 5 to find the top element
Enter any other number to exit\n"""))  # User menu

    if user_input == 1:
        value = int(input("Enter the value to push: "))
        stack = Push(stack, value, stack_size)
        print("Stack:", stack)
    elif user_input == 2:
        stack = Pop(stack)
        print("Stack:", stack)
    elif user_input == 3:
        isFull(stack, stack_size)
    elif user_input == 4:
        isEmpty(stack)
    elif user_input == 5:
        findTop(stack)
    else:
        break
```

**Implementation of the “Queue using an array”. Use these concept: Enqueue, Dequeue, Front, Rear, isfull() and isempty() **

```python
# Function to enqueue an element into the queue
def enqueue(queue, value, queue_size, rear, front):
    if rear == queue_size - 1:  # Check if the queue is full #Rear is at end position
        print("Queue Overflow")
    else:
        if front == -1:  # If the queue is empty, set front to 0
            front = 0 #only for first element

        rear += 1  # Move the rear pointer to the next position
        queue.append(value)  # Add the value to the queue
    return queue, rear, front

# Function to dequeue an element from the queue
def dequeue(queue, front, rear):
    if front == -1 :  # Check if the queue is empty
        print("Queue Underflow")
    else:
        value = queue[front]  # Get the value from the front of the queue
        queue[front] = None  # Remove the value from the queue
        front += 1  # Move the front pointer to the next position
        if front > rear:  # If the queue becomes empty after dequeue #when F and R pointing to same element before dequeue
            front = rear = -1  # Reset both front and rear pointers
        return queue, front, rear, value

# Function to get the front pointer of the queue
def get_front(front):
    if front == -1:  # If the queue is empty
        print("Queue is empty, no front")
    else:
        print("Front pointer:", front)

# Function to get the rear pointer of the queue
def get_rear(rear):
    if rear == -1:  # If the queue is empty
        print("Queue is empty, no rear")
    else:
        print("Rear pointer:", rear)

# Function to check if the queue is full
def is_full(queue_size, rear):
    if rear == queue_size - 1:  # Check if the rear pointer is at the last position
        print("The queue is full")
    else:
        print("The queue is not full")

# Function to check if the queue is empty
def is_empty(front):
    if front == -1:  # If the front pointer is -1, the queue is empty
        print("The queue is empty")
    else:
        print("The queue is not empty")


# Main code
queue_size = int(input("Enter the size of the queue: "))
front, rear = -1, -1  # Initialize front and rear pointers
queue = []  # Initialize an empty queue

while True:
    # User menu
    user_input = int(input("""Enter 1 to enqueue
Enter 2 to dequeue
Enter 3 to print the front pointer
Enter 4 to print the rear pointer
Enter 5 to check whether the queue is full or not
Enter 6 to check whether the queue is empty or not
Enter any other number to exit\n"""))

    if user_input == 1:
        value = int(input("Enter the value to enqueue: "))
        queue, rear, front = enqueue(queue, value, queue_size, rear, front)
        print("Queue:", queue)
    elif user_input == 2:
        queue, front, rear, dequeued_value = dequeue(queue, front, rear)
        if dequeued_value is not None:
            print("Dequeue Element:", dequeued_value)
        print("Queue:", queue)
    elif user_input == 3:
        get_front(front)
    elif user_input == 4:
        get_rear(rear)
    elif user_input == 5:
        is_full(queue_size, rear)
    elif user_input == 6:
        is_empty(front)
    else:
        break
```

**C.Implementation of the “Circular Queue using an array” **

```python
# Function to enqueue an element into the queue
def enqueue(queue, value, size, front, rear):
    # Check if the next position of rear is equal to front, indicating queue overflow
    if (rear + 1) % size == front:
        print("Queue Overflow")
    # If queue is empty, set front and rear to 0 and add the value at rear
    elif front == -1:
        front = rear = 0
        queue[rear] = value #Elements added at Rear End
    # Otherwise, move rear to the next position and add the value
    else:
        rear = (rear + 1) % size
        queue[rear] = value
    return queue, front, rear

# Function to dequeue an element from the queue
def dequeue(queue, size, front, rear):
    # Check if the queue is empty
    if front == -1:
        print("Queue Underflow")
    # If only one element is left in the queue
    elif front == rear:
        print("Dequeued Value:", queue[front])
        front = rear = -1
    # Otherwise, move front to the next position
    else:
        print("Dequeued Value:", queue[front])
        front = (front + 1) % size
    return queue, front, rear

# Function to display the elements of the queue
def display(queue, size, front, rear):
    # Check if the queue is empty
    if front == -1:
        print("Queue is empty")
    # If elements are contiguous in memory
    elif rear >= front:
        print("Queue:", end=" ")
        for i in range(front, rear + 1):
            print(queue[i], end=" ")
        print()
    # If elements wrap around the end of the array
    else:
        print("Queue wa:", end=" ")
        for i in range(front, size):
            print(queue[i], end=" ")
        for i in range(0, rear + 1):
            print(queue[i], end=" ")
        print()
def get_front(front):
    if front == -1:  # If the queue is empty
        print("Queue is empty, no front")
    else:
        print("Front pointer:", front)

# Function to get the rear pointer of the queue
def get_rear(rear):
    if rear == -1:  # If the queue is empty
        print("Queue is empty, no rear")
    else:
        print("Rear pointer:", rear)
# Example usage:
queue_size = int(input("Enter the size of the queue: "))
queue = [None] * queue_size
front = rear = -1

# Main loop for user interaction
while True:
    user_input = int(input("""Enter 1 to enqueue
Enter 2 to dequeue
Enter 3 to display the queue
Enter 4 to display front pointer
Enter 5 to display rear pointer
Enter any other number to exit\n"""))

    if user_input == 1:
        value = int(input("Enter the value to enqueue: "))
        queue, front, rear = enqueue(queue, value, queue_size, front, rear)
        display(queue, queue_size, front, rear)
    elif user_input == 2:
        queue, front, rear = dequeue(queue, queue_size, front, rear)
    elif user_input == 3:
        display(queue, queue_size, front, rear)
    elif user_input == 4:
        get_front(front)
    elif user_input == 5:
        get_rear(rear)
    else:
        break
```

**No. 04 - Practice Problems: Merge Sort and Quick Sort **

Write a program for implementing the following sorting techniques to arrange a list of integers in ascending order.

A)Quick sort
B)Merge Sort

```python
#Solution for Quick sort:

def partition(array, low, high):
    pivot = array[high]  # Choosing the last element as the pivot
    i = low - 1  # Index of the smaller element

    for j in range(low, high):
        if array[j] < pivot:
            # If current element is smaller than the pivot
            # swap arr[i] and arr[j]
            i = i + 1
            array[i], array[j] = array[j], array[i]

    # swap arr[i+1] and arr[high] (or pivot)
    array[i + 1], array[high] = array[high], array[i + 1]
    return i + 1

def quicksort(array, low, high):
    if low < high:
        # pi is partitioning index, array[p] is now at right place
        pi = partition(array, low, high)

        # Separately sort elements before partition and after partition
        quicksort(array, low, pi - 1)
        quicksort(array, pi + 1, high)

if __name__ == '__main__':
    array = [9, 1, 2, 3, 4, 5]
    N = len(array)

    quicksort(array, 0, N - 1)
    print("sorted array:",array)
```

```python
# Input array
array = [10, 1, 9, 7, 5, 4, 3, 2, 6]

# Print original array
print("Original array:", array)

# Merge sort function
def mergeSort(array):
    # Check if the array has more than one element
    if len(array) > 1:
        # Find the middle index
        mid = len(array) // 2

        # Divide the array into two halves
        L = array[:mid]  # Left half
        R = array[mid:]  # Right half

        # Recursive calls to mergeSort for each half
        mergeSort(L)
        mergeSort(R)

        # Initialize pointers for L, R, and main array
        i = j = k = 0

        # Merge the sorted halves back into the main array
        while i < len(L) and j < len(R):
            if L[i] <= R[j]:
                array[k] = L[i]
                i += 1
            else:
                array[k] = R[j]
                j += 1
            k += 1

        # Insert remaining elements of L (if any)
        while i < len(L):
            array[k] = L[i]
            i += 1
            k += 1

        # Insert remaining elements of R (if any)
        while j < len(R):
            array[k] = R[j]
            j += 1
            k += 1

# Call mergeSort function
mergeSort(array)

# Print sorted array
print("Sorted array:", array)
```
